```python
import torch
import matplotlib.pyplot as plt
```

```python
def make_mean_zero(X):
    X -= torch.mean(X,0,True)
    return X
def covariance(X):
    X_t=torch.transpose(X,0,1)
    #print(X_t)
    covx=torch.mm(X_t,X)/X.size()[0]
    #print(covx)
    return covx
```

```python
X=torch.load("assign0_data.py")
plt.rc('figure',figsize=(6,6))
plt.scatter(X[:,0].numpy(),X[:,1].numpy())
plt.savefig('dataplot_raw')
plt.close()
X=make_mean_zero(X)
#print(torch.mean(X,-1,keepdim=True))
#plt.matshow(X.numpy())
#plt.show()
plt.rc('figure',figsize=(6,6))
plt.scatter(X[:,0].numpy(),X[:,1].numpy())
plt.savefig('dataplot_mean0')
plt.close()

covx=covariance(X)
U,S,V = torch.svd(covx)
Xrot = torch.mm(X, U)
Xwhite = Xrot / torch.sqrt(S + 1e-5)
plt.rc('figure',figsize=(6,6))
plt.scatter(Xwhite[:,0].numpy(),Xwhite[:,1].numpy())
plt.savefig('dataplot_w')
plt.close()
```
